You can use train_test_split() to solve classification problems the same way you do for regression analysis.

```python
from sklearn.model_selection import train_test_split
x= data.drop("Salary",axis=1)
```

```python
x.head()
```

```python
y = data["Salary"]
```

```python
y.head()
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(x,y,test_size=0.2,random_state=42)
```

# ***Linear*** ***Regression***

 Linear Regression is a supervised learning algorithm for predicting continuous values based on input variables. This algorithm establishes a linear relationship between the independent variables (input variables, features, or predictors) and the dependent variable (output variable or target variable)

```python
from sklearn.linear_model import LinearRegression
LR =LinearRegression()
```

```python
LR.fit(xtrain,ytrain)
```

```python
y_predection = LR.predict(xtest)
```

```python
ytest
```

```python
y_predection
```

```python
print(LR.score(xtest,ytest))
```

```python
avg =(LR.score(xtest,ytest))
```
