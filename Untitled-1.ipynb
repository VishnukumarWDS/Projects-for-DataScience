```python
from sklearn.feature_selection import mutual_info_regression

# Assuming the target variable is 'target_column'
X = data.drop(columns=[target_column])
y = data[target_column]

# Calculate mutual information
mi = mutual_info_regression(X, y)

mi_series = pd.Series(mi, index=X.columns)
mi_series = mi_series.sort_values(ascending=False)

plt.figure(figsize=(12, 6))
mi_series.plot(kind='bar')
plt.title('Mutual Information of Features for Regression')
plt.show()
```

```python
from sklearn.cluster import KMeans

# Perform K-means clustering
kmeans = KMeans(n_clusters=3)  # Assuming we want 3 clusters
data['cluster'] = kmeans.fit_predict(X)

plt.figure(figsize=(10, 6))
sns.scatterplot(x=X.columns[0], y=X.columns[1], hue='cluster', data=data, palette='viridis')
plt.title('K-Means Clustering')
plt.show()
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)

# Convert back to DataFrame for ease of use
X_scaled = pd.DataFrame(X_scaled, columns=X.columns)

# Display the first few rows of the scaled dataset
print(X_scaled.head())
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# Split the data
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# Train a baseline model
model = LinearRegression()
model.fit(X_train, y_train)

# Predictions
y_pred = model.predict(X_test)

# Evaluation
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f'Mean Squared Error: {mse}')
print(f'R-squared: {r2}')
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestRegressor

# Define the parameter grid
param_grid = {
    'n_estimators': [50, 100, 200],
    'max_depth': [None, 10, 20, 30],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}

# Initialize the model
rf = RandomForestRegressor(random_state=42)

# Initialize GridSearchCV
grid_search = GridSearchCV(estimator=rf, param_grid=param_grid, cv=3, n_jobs=-1, verbose=2)

# Fit the model
grid_search.fit(X_train, y_train)

# Best parameters and model
print(f'Best Parameters: {grid_search.best_params_}')
best_model = grid_search.best_estimator_

# Predictions with the best model
y_pred_best = best_model.predict(X_test)

# Evaluation
mse_best = mean_squared_error(y_test, y_pred_best)
r2_best = r2_score(y_test, y_pred_best)

print(f'Mean Squared Error (Best Model): {mse_best}')
print(f'R-squared (Best Model): {r2_best}')
```
